```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/Users/gayatridevadula/Downloads/Salary.csv")
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='YearsExperience', y='Salary', data=df)
plt.title('Years of Experience vs. Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='YearsExperience', y='Salary', data=df)
plt.title('Salary Over Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Salary'], bins=10, kde=True)
plt.title('Distribution of Salary')
plt.xlabel('Salary')
plt.ylabel('Frequency')
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Salary', data=df, orient='v')
plt.title('Boxplot of Salary')
```

```python
sns.pairplot(df)
```

```python
sns.jointplot(x='YearsExperience', y='Salary', data=df, kind='reg')
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(df['Salary'], fill=True)
plt.title('Kernel Density Estimation (KDE) of Salary')
```

```python
average_salary = df.groupby('YearsExperience')['Salary'].mean()
average_salary.plot(kind='bar', figsize=(10, 6))
plt.title('Average Salary per Year of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Average Salary')
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='YearsExperience', y='Salary', data=df)
plt.title('Linear Regression Plot of Salary vs. Years of Experience')
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Salary', data=df, orient='v', inner='quartile')
plt.title('Violin Plot of Salary Distribution')
```
